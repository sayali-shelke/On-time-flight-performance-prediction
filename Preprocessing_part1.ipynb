```python
data14 = pd.DataFrame(df14,columns=['ARR_TIME','CRS_ARR_TIME','DEP_TIME','CRS_DEP_TIME'])
data14 = pd.to_numeric(data14.stack(), errors='coerce').fillna(0)
df2= (pd.to_timedelta(data14 // 100, unit='hours') + pd.to_timedelta(data14 % 100, unit='T')).unstack()
ddf = df14.dropna(axis=0, subset=['ARR_TIME','CRS_ARR_TIME','DEP_TIME','CRS_DEP_TIME'])
df_merged14 = ddf.merge(df2, how='outer', left_index=True, right_index=True)
```

```python
l2=['ATL','DEN','CLT','DFW','EWR']
df144=df_merged14[df_merged14['ORIGIN'].isin(l2)]

df144=df144[df144['DEST'].isin(l2)]
```

```python
df144.drop(labels=['Unnamed: 0','Unnamed: 21','Unnamed: 23','ARR_TIME_x','CRS_ARR_TIME_x','DEP_TIME_x','CRS_DEP_TIME_x'],inplace=True,axis=1)
```

```python
df144['FL_DATE']=pd.to_datetime(df144['FL_DATE'])
```

```python
df_merged14=df144[pd.notnull(df144["ARR_TIME_y"])]
```

```python
df_merged14["ARR_DateTime"] = df_merged14["FL_DATE"] + df_merged14["ARR_TIME_y"].map(pd.Timedelta.to_pytimedelta)
```

```python
df_merged14["DEP_DateTime"] = df_merged14["FL_DATE"] + df_merged14["DEP_TIME_y"].map(pd.Timedelta.to_pytimedelta)
```

```python
df_merged14.loc[df_merged14['ARR_TIME_y'] > df_merged14['CRS_ARR_TIME_y'], 'target'] = 1
df_merged14.loc[df_merged14['ARR_TIME_y'] <= df_merged14['CRS_ARR_TIME_y'], 'target'] = 0
```
